# Exploratory Data Analysis
start by exploring the response variable grade, the one we are supposed to predict at the end of the day

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

data = pd.read_csv('student-por.csv')
```

# categorical variables 
- address (address type of student, rural or urban)
- paid (presence of extra tuition classes, yes or no)
- activities/extra-curriculars (yes or no)
- wants higher education (intention to pursue higher education, yes or no)
- reason to choose this school (home(school is close to home), reputation(school's reputation), course(school has preferred course of student) ,other)

variable 1: address (rural or urban)

```python
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

combined = pd.DataFrame(data[['G3','address']])
sb.boxplot(x='address',y='G3',data=combined)
```

Urban has a higher median score and a larger range of scores

```python
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

address = pd.DataFrame(data['address'])   # Response
G3 = pd.DataFrame(data['G3'])     # Predictor

# Split the Dataset into Train and Test
X1_train, X1_test, y1_train, y1_test = train_test_split(G3, address, test_size = 0.20, random_state=4)

# First Decision Tree using Train Data with max_depth 3
dectree1 = DecisionTreeClassifier(max_depth = 3)  # create the decision tree object
dectree1.fit(X1_train, y1_train)                    # train the decision tree model

# Plot the trained Decision Tree
from sklearn.tree import plot_tree

f = plt.figure(figsize=(30,30))
plot_tree(dectree1, filled=True, rounded=True, 
          feature_names=["G3"], 
          class_names=["Rural","Urban"])

y1_train_pred = dectree1.predict(X1_train)
y1_test_pred = dectree1.predict(X1_test)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree1.score(X1_train, y1_train))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", dectree1.score(X1_test, y1_test))
print()

print("Train Set on the left")
print("Test Set on the right")

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y1_train, y1_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y1_test, y1_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
```
